FINE TUNING

```python
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
#AutoTokenizer is a tokenizer utility from Hugging Face’s transformers library. Tokenizers convert text into token IDs, which can be processed by models. AutoTokenizer allows you to load a pre-trained tokenizer for the specific model you’re using.
#AutoModelForSeq2SeqLM is a sequence-to-sequence (Seq2Seq) model architecture also from the transformers library. It's useful for generating a sequence from a given input sequence, as in tasks like translation, summarization, and question-answering.
from peft import LoraConfig, get_peft_model
#LoraConfig configures the settings for Low-Rank Adaptation (LoRA), a PEFT technique that reduces memory and computational overhead during fine-tuning.
#get_peft_model applies PEFT settings to your model, helping you efficiently fine-tune a model without extensive resources, making it ideal for tasks with limited data or computational power.
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments, DataCollatorForSeq2Seq
# Seq2SeqTrainer is a specialized Trainer class for Seq2Seq tasks, streamlining the process of training these models.
# Seq2SeqTrainingArguments holds configuration options for the training process, such as batch size, learning rate, and number of epochs.
# DataCollatorForSeq2Seq is a utility for collating data into batches, padding sequences as needed to match the model’s input requirements.
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(ml_df, test_size=0.1, random_state=0, shuffle=True)
train_data = Dataset.from_pandas(train_df)
test_data = Dataset.from_pandas(test_df)
```

```python
model_id="google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_id) #  The from_pretrained() method automatically finds and sets up the tokenizer associated with the model ID you provided.
model = AutoModelForSeq2SeqLM.from_pretrained(model_id) # The from_pretrained method is used to load a model that has already been trained (a pre-trained model) and is ready for use.
```

# FlanT5
"google/flan-t5-large" is the name of a particular model, FLAN-T5 Large, which is fine-tuned for tasks like summarization, translation, and question answering.
FLAN-T5 is a variant of the T5 (Text-To-Text Transfer Transformer) model developed by Google, and the "large" version has a high number of parameters for enhanced performance.

# Working of preprocess_function
This preprocess_function tokenizes and formats the questions and answers in the dataset to make them ready for model training. By padding, truncating, and adjusting the labels, it ensures that data inputs are consistent in length, and padding tokens in the labels are ignored during loss calculation.

For each token l in each label (which is an inner list), the code checks if l is equal to tokenizer.pad_token_id, which is the ID representing a padding token.
If l (the token) is not equal to tokenizer.pad_token_id, it’s kept as-is (l).
If l is equal to tokenizer.pad_token_id, it’s replaced with -100.

```python
def preprocess_function(sample,padding="max_length"):
    model_inputs = tokenizer(sample["Questions"], max_length=256, padding=padding, truncation=True)
    labels = tokenizer(sample["Answers"], max_length=128, padding=padding, truncation=True)
    if padding == "max_length":
        labels["input_ids"] = [
            [(l if l != tokenizer.pad_token_id else -100) for l in label] for label in labels["input_ids"]
        ] # tokenizer.pad_token_id is the ID used for padding, and replacing it with -100 helps the model ignore padding tokens when calculating training loss, focusing only on the meaningful parts of the sequence.
    model_inputs["labels"] = labels["input_ids"] #adds the processed labels (the tokenized answer sequence) to model_inputs under the key "labels".
    return model_inputs
```

**Explaination :** Tokenizes both the training and test datasets.
Removes the original columns (like "Questions" and "Answers") after tokenization.
Prints out the keys of the tokenized dataset to confirm the fields available for model input. This ensures the data is correctly formatted for training and testing.

```python
train_tokenized_dataset = train_data.map(preprocess_function, batched=True, remove_columns=train_data.column_names)
test_tokenized_dataset = test_data.map(preprocess_function, batched=True, remove_columns=test_data.column_names)
print(f"Keys of tokenized dataset: {list(train_tokenized_dataset.features)}")
```

```python
lora_config = LoraConfig(
 r=8,
 lora_alpha=16,
 lora_dropout=0.1,
 bias="none",
 task_type="SEQ_2_SEQ_LM",
 target_modules=["q", "v"]
)
```

```python
model
```

```python
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
model
```

```python
label_pad_token_id = -100
data_collator = DataCollatorForSeq2Seq(
    tokenizer,
    model=model,
    label_pad_token_id=label_pad_token_id,
    pad_to_multiple_of=8
)
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
output_dir="flant5-large-aio"
training_args = Seq2SeqTrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=6,
    learning_rate=2e-4,
    num_train_epochs=1,
    logging_dir=f"{output_dir}/logs",
    logging_strategy="epoch",
    save_strategy="epoch",
    report_to="tensorboard",
    push_to_hub = True
)
```

```python
model.config.use_cache = False
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_tokenized_dataset
)
```

```python
trainer.train()
```

```python
peft_save_model_id="flant5-large-aio"
trainer.model.save_pretrained(peft_save_model_id, push_to_hub=True)
tokenizer.save_pretrained(peft_save_model_id, push_to_hub=True)
trainer.model.base_model.save_pretrained(peft_save_model_id, push_to_hub=True)
```
